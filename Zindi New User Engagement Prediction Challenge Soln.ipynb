```python
def get_latest_interaction_date(userid, df):
    """
    This function will give us the latest interaction of a user
    """
    
    dat = df.copy()
    
    x = dat[dat['User_ID'] == userid]
#     print(x)
    try:
        return max(x['Time_Comment'])
    except:
        return pd.to_datetime('01-01-1920')
```

```python
Users['Latest_Time_Comments'] = Users.apply(lambda x: get_latest_interaction_date(x['User_ID'], Comments), axis = 1)
```

```python
Users['Latest_Time_Discussion'] = Users.apply(lambda x: get_latest_interaction_date(x['User_ID'], Discussion), axis = 1)
```

```python
Users['Latest_Time_Comp'] = Users.apply(lambda x: get_latest_interaction_date(x['User_ID'], CompetitionParticipation), axis = 1)
```

```python
Users['Max_Time'] = Users.apply(lambda x: max(x['Latest_Time_Comments'], x['Latest_Time_Discussion'], x['Latest_Time_Comp']), axis = 1)
```

```python
Users['Time_Created'].dt.month
```

```python
Users['Target'] = Users.apply(lambda x: 1 if (x['Max_Time'] - x['Time_Created']).days >= 30 else 0, axis = 1)
```
